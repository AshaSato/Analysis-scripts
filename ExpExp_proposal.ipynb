## Background stuff

Lee & Navarro (2009) presented people with point-patterns based on real constellations and ask them to join up the points such that all are included, but no other restrictions. They found significant inter-participant overlap/agreement in the resulting structures, and signnificant overlap in the structures with a subset of the Delauney triangulation, in particular the minimum spanning tree (MST) (86% overlap). This suggests a bias in connecting point patterns towards producing structures that do so efficiently.

### Training

Training is identical in both conditions: Participants see a star map without a label, are asked to guess the meaning (forward guessing to improve recall), and are then given the correct target label. The only difference between conditions are the correspondence between starmaps and labels (as described above). 
### Testing

#### guessing task
Participants see a star map and are asked to guess its label (free guess? or pick from an array?). 

Prediction: people will be better at guessing in the iconic condition.

#### connecting task

Participants are asked to connect the points. I think it's important that this is done before giving them the meaning?

#### rating task
Participants are given the target label (thereby receiving feedback on their guess) and are asked to give an iconicity rating, e.g. "How much do you think this looks like a _____".
